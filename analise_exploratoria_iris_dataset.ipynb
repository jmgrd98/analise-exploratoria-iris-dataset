# Análise exploratória do famoso dataset de flores Íris

Este é o famoso dataset de flores Iris utilizado foi baixado no site do Kaggle: https://www.kaggle.com/datasets/uciml/iris

O dataset inclui três espécies de íris com 50 amostras cada, bem como algumas propriedades de cada flor. Uma espécie de flor é linearmente separável das outras duas, mas as outras duas não são linearmente separáveis ​​uma da outra.

Primeiro importamos a base utilizando a biblioteca Pandas e visualizamos os 5 primeiros registros.

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/datasets/Iris.csv')

df.head()
```

Nesse dataset foram registradas igualmente 50 representantes de cada uma das 3 sub-espécies de flores íris.

```python
qty_species = df['Species'].value_counts()
print(qty_species)

species = ['Iris-virginica', 'Iris-versicolor', 'Iris-setosa']
plt.pie(qty_species, labels=species, colors=['red', 'green', 'blue'], autopct='%1.1f%%')
```

Visualizando as flores com maiores sépalas em comprimento (das maiores para as menores) podemos verificar que as 5 maiores são da espécie *Iris-virginica* e as 5 menores são da espécie *Iris-setosa*.

```python
df.sort_values('SepalLengthCm', ascending=False)
```

Abaixo podemos ver a média do comprimento das sépalas de ambas as espécies.

```python
media_virginica = df.loc[df['Species'] == 'Iris-virginica', 'SepalLengthCm'].mean().round(2)
media_setosa = df.loc[df['Species'] == 'Iris-setosa', 'SepalLengthCm'].mean().round(2)

print('Média virginica: ', media_virginica)
print('Média setosa: ', media_setosa)
```

E assim constatamos que a média do comprimento das sépalas da espécie *Iris-virginica* é de fato maior do que a média da espécie *Iris-setosa*.

Para visualizarmos melhor, vamos plotar um gráfico com as médias das 3 espécies.

Agora calculamos a média das 3 espécies e plotamos em um gráfico de barras utilizando a biblioteca Matplotlib.

```python
import matplotlib.pyplot as plt

media_virginica = df.loc[df['Species'] == 'Iris-virginica', 'SepalLengthCm'].mean().round(2)
media_setosa = df.loc[df['Species'] == 'Iris-setosa', 'SepalLengthCm'].mean().round(2)
media_versicolor = df.loc[df['Species'] == 'Iris-versicolor', 'SepalLengthCm'].mean().round(2)

species = ['Iris-virginica', 'Iris-versicolor', 'Iris-setosa']
means = [media_virginica, media_versicolor, media_setosa]

plt.bar(species, means, color=['blue', 'green', 'red'])

plt.title('Média do comprimento de sépalas de cada espécie')
plt.xlabel('Espécie')
plt.ylabel('Comprimento (cm)')

plt.show()
```
